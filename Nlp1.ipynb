```python
chapterl_file = open("chapterstext.txt",encoding='utf-8',errors='ignore')


chapterl_wordslist = chapterl_file.read().splitlines() # storing lines of novel in wordlist
chapterl_wordslist = [i for i in chapterl_wordslist if i!=''] #removing empty lines from wordlist
chapterl_text_string = ""
chapterl_text_string = chapterl_text_string.join(chapterl_wordslist)  #concatenating all lines in the string
chapterl_text_string= re.sub(pattern, '', chapterl_text_string)

chapterl_text_string
```

```python
import string
punctuation_set = set(string.punctuation)
chapterl_processed_text = ""
for c in chapterl_text_string:
    if c not in punctuation_set:
        chapterl_processed_text = chapterl_processed_text + c

#Converting the text into lower case
chapterl_processed_text = chapterl_processed_text.lower()
chapterl_processed_text
```

```python
chapterl_words = nltk.word_tokenize(chapterl_processed_text)
bi_grams = list(bigrams(chapterl_words))
bigram_freq = FreqDist(bi_grams)
bigram_probabilities = {}
for bigram in bigram_freq:
    first_word = bigram[0]
    probability = bigram_freq[bigram] / chapterl_words.count(first_word)
    bigram_probabilities[bigram] = probability
bigram_probabilities = sorted(bigram_probabilities.items(), key=lambda x: x[1], reverse=True) # storing biagrams in descending order of probability
```

```python
for bigram, probability in bigram_probabilities:
    print(f"{bigram[0]} {bigram[1]}: {probability}")
```
